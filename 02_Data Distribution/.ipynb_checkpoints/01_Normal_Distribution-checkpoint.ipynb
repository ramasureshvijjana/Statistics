# Normal Distribution

```python
# Importing the libreries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Making data
data = pd.Series([23, 25, 20, 27, 22, 26, 24, 23, 24, 
                  25, 21, 23, 24, 25, 28, 23, 24, 25, 
                  23, 25, 20, 27, 22, 19, 26, 24, 21])
data = data.sort_values()
```

```python
sns.kdeplot(color=)

```

```python
sns.histplot(data = data, discrete = "false", kde=True, color="skyblue", edgecolor='pink', linestyle='--')

# Add title and labels
plt.title('Histogram with Line Plot Overlay')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```
